```python
import matplotlib.pyplot as plt
from skimage.io import imread,imshow
```

```python
im = imread("lena.jpg")
```

```python
imshow(im)
```

```python
print(type(im))
```

```python
import numpy as np
# creating a lookup table (LUT)
LUT = np.array([255-i for i in range(256)])
print(LUT)
```

```python
plt.figure()
plt.plot(range(256),LUT)
plt.show()
```

```python

# Apply LUT
im_out = LUT[im]
plt.figure()
plt.imshow(im_out,plt.cm.gray)
plt.show()
```
